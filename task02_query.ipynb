# **Nhiệm vụ 2 - Truy vấn & thống kê mô tả**
*Mục tiêu : thực hành truy xuất dữ liệu và rút ra các thống kê cơ bản*

```python
# import libary
import pandas as pd

# Read csv
df_teams = pd.read_csv(f'./data_clean/team_info_clean.csv')
df_players = pd.read_csv(f'./data_clean/player_stats_clean.csv')
df_matches = pd.read_csv(f'./data_clean/match_results_clean.csv')
```

**1. Thống kê số đội bóng theo home_city**

```python
team_count = df_teams.groupby('home_city').size().reset_index(name='team_count')
team_count
```

**2. Thống kê số cầu thủ theo mỗi đội bóng**

```python
player_count = df_players.groupby('team_id').size().reset_index(name='player_count')
player_count = player_count.merge(
    df_teams[['team_id', 'team_name']],
    on='team_id',
    how='left'
)
player_count
```

**3. Danh sách cầu thủ ghi được nhiều hơn một số bàn thắng nhất định (>=5)**

```python
player_with_team = df_players.merge(
    df_teams[['team_id', 'team_name']],
    on='team_id',
    how='left'
)
top_scorers = player_with_team[player_with_team['goals'] >= 5][
    ['player_id', 'player_name', 'team_name', 'goals']
].sort_values('goals', ascending=False)
top_scorers
```

**4. Thống kê số trận đấu đã diễn ra trong mùa giải theo tháng**

```python
matches_per_month = (
    df_matches.groupby(pd.to_datetime(df_matches['match_date']).dt.to_period('M'))
        .size()
        .reset_index(name='match_count')
        .rename(columns={'match_date': 'month'})
)
matches_per_month
```

**5. Các trận đấu có tỷ số cách biệt lớn nhất**

```python
df_matches['hieu_so'] = (df_matches['home_goals'] - df_matches['away_goals']).abs()
max_diff = df_matches['hieu_so'].max()
matches = df_matches[df_matches['hieu_so'] == max_diff]
matches
```
